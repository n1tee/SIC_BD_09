```python
import pandas as pd
import requests
import time
import random
from tqdm import tqdm
import os

df_id = pd.read_csv("product_id.csv")
p_ids = df_id["id"].to_list()

start_index = 698
p_ids_to_crawl = p_ids[start_index:]
print(f"Tiếp tục crawl từ vị trí {start_index} – ID: {p_ids[start_index]}. Còn lại: {len(p_ids_to_crawl)} sản phẩm.")

result = []

for pid in tqdm(p_ids_to_crawl, desc="Crawling sản phẩm tiếp tục", total=len(p_ids_to_crawl)):
    try:
        response = requests.get(
            f"https://tiki.vn/api/v2/products/{pid}",
            headers=headers,
            params=params,
            cookies=cookies
        )
        if response.status_code == 200:
            result.append(parser_product(response.json()))
        else:
            print(f"Lỗi khi lấy sản phẩm {pid}: {response.status_code}")
    except Exception as e:
        print(f"Lỗi không xác định với sản phẩm {pid}: {e}")
    
    time.sleep(random.randint(3, 5))

df_product_new = pd.DataFrame(result)

output_file = "product_data.csv"
if os.path.exists(output_file):
    df_existing = pd.read_csv(output_file)
    df_combined = pd.concat([df_existing, df_product_new], ignore_index=True)
    df_combined.drop_duplicates(subset=['id'], inplace=True)
    df_combined.to_csv(output_file, index=False)
    print(f"Đã nối dữ liệu vào {output_file}. Tổng số sản phẩm: {len(df_combined)}")
else:
    df_product_new.to_csv(output_file, index=False)
    print(f"Đã tạo mới {output_file} và lưu dữ liệu. Tổng sản phẩm: {len(df_product_new)}")
```
